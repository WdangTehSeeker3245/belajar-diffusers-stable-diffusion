```python
import os

# Specify the directory path where you want to delete .png files
directory_path = "/content"

# Iterate through the files in the directory
for filename in os.listdir(directory_path):
    if filename.endswith(".png"):
        file_path = os.path.join(directory_path, filename)
        os.remove(file_path)
        print(f"Deleted: {file_path}")
```

```python
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("faizalnf1800/StingerMix-v4.0-SD1.5")
```

```python
pipe.load_lora_weights("faizalnf1800/sidebags-earring-anime-woman-lora")
pipe.to("cuda")
pipe.safety_checker = None
pipe.requires_safety_checker = False
```

```python
prompt = "A woman have long black hair with side bags hair style and wear earring."
image = pipe(prompt, num_inference_steps=30, guidance_scale=7.5).images[0]
image.save("image.png")
```

```python
from huggingface_hub import upload_folder

output_dir = "/content/sidebags-earring-anime-woman-lora"
hub_model_id = "faizalnf1800/sidebags-earring-anime-woman-lora"

upload_folder(output_dir, hub_model_id)
```

```python
import torch

# Convert the model to safetensors format
pipe.to_safetensors()

# Save the safetensors file
torch.save(pipe, "sidebags-earring-anime-woman-lora.safetensors")
```
